```python

def simple_ner(text):
    tokens = text.split()
    results = []

    for token in tokens:

        clean_token = re.sub(r'[^\w/:-]', '', token)

        normalized = clean_token.title()

        # PERSON
        if normalized in PERSON_LIST:
            results.append((clean_token, "PERSON"))

        # LOCATION
        elif normalized in LOCATION_LIST:
            results.append((clean_token, "LOCATION"))

        # ORGANIZATION
        elif normalized in ORG_LIST:
            results.append((clean_token, "ORGANIZATION"))

        # Days of the week
        elif normalized in DAYS:
            results.append((clean_token, "DATE/TIME"))

        # Months
        elif normalized in MONTHS:
            results.append((clean_token, "DATE/TIME"))

        # Dates (e.g., 5/7/2026 or 05/07/26)
        elif re.fullmatch(r'\d{1,2}/\d{1,2}/\d{2,4}', clean_token):
            results.append((clean_token, "DATE/TIME"))

        # Times (e.g., 10:30, 9:15)
        elif re.fullmatch(r'\d{1,2}:\d{2}', clean_token):
            results.append((clean_token, "DATE/TIME"))

    return results
```

```python
def evaluate_ner(expected, predicted):

    total_tp = 0
    total_fp = 0
    total_fn = 0

    for exp_par, pred_par in zip(expected, predicted):

        # convert to sets
        exp_set = set(exp_par)
        pred_set = set(pred_par)

        # exact match evaluation
        tp = len(exp_set & pred_set)
        fp = len(pred_set - exp_set)
        fn = len(exp_set - pred_set)

        total_tp += tp
        total_fp += fp
        total_fn += fn

    precision = total_tp / (total_tp + total_fp) if (total_tp + total_fp) else 0
    recall = total_tp / (total_tp + total_fn) if (total_tp + total_fn) else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) else 0

    print("\n===== FINAL NER EVALUATION =====\n")
    print(f"TP: {total_tp}")
    print(f"FP: {total_fp}")
    print(f"FN: {total_fn}")
    print("--------------------------------")
    print(f"Precision: {precision:.2%}")
    print(f"Recall   : {recall:.2%}")
    print(f"F1 Score : {f1:.2%}")
    print("--------------------------------\n")
```

```python
def normalize(text):
    return text.lower().replace(".", "").replace(",", "").strip()
```

```python
def match_entities(exp, pred):
    exp = [normalize(x[0]) for x in exp]
    pred = [normalize(x[0]) for x in pred]

    tp = 0
    fp = 0
    fn = 0

    for p in pred:
        if any(p in e or e in p for e in exp):
            tp += 1
        else:
            fp += 1

    for e in exp:
        if not any(e in p or p in e for p in pred):
            fn += 1

    return tp, fp, fn
```

```python
def evaluate_fixed(expected, predicted):
    tp = fp = fn = 0

    for exp, pred in zip(expected, predicted):
        t_tp, t_fp, t_fn = match_entities(exp, pred)
        tp += t_tp
        fp += t_fp
        fn += t_fn

    precision = tp / (tp + fp + 1e-9)
    recall = tp / (tp + fn + 1e-9)
    f1 = 2 * precision * recall / (precision + recall + 1e-9)

    print("\n===== FINAL FIXED EVALUATION =====")
    print("TP:", tp, "| FP:", fp, "| FN:", fn)
    print("--------------------------------")
    print("Precision:", round(precision * 100, 2))
    print("Recall   :", round(recall * 100, 2))
    print("F1 Score :", round(f1 * 100, 2))
    print("--------------------------------")
```

```python
def normalize(text):
    return text.lower().replace(".", "").replace(",", "").strip()


def match_entities(exp, pred):
    exp = [normalize(x[0]) for x in exp]
    pred = [normalize(x[0]) for x in pred]

    tp = fp = fn = 0

    for p in pred:
        if any(p in e or e in p for e in exp):
            tp += 1
        else:
            fp += 1

    for e in exp:
        if not any(e in p or p in e for p in pred):
            fn += 1

    return tp, fp, fn


def evaluate_test(expected, predicted):
    tp = fp = fn = 0

    for exp, pred in zip(expected, predicted):
        t_tp, t_fp, t_fn = match_entities(exp, pred)
        tp += t_tp
        fp += t_fp
        fn += t_fn

    precision = tp / (tp + fp + 1e-9)
    recall = tp / (tp + fn + 1e-9)
    f1 = 2 * precision * recall / (precision + recall + 1e-9)

    print("\n===== FINAL TEST RESULT =====")
    print("TP:", tp)
    print("FP:", fp)
    print("FN:", fn)
    print("----------------------------")
    print("Precision:", round(precision * 100, 2), "%")
    print("Recall   :", round(recall * 100, 2), "%")
    print("F1 Score :", round(f1 * 100, 2), "%")
    print("----------------------------")


evaluate_test(expected, predicted)
```
