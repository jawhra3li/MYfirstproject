```python

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model


(x_train, _), (x_test, _) = tf.keras.datasets.mnist.load_data()

x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Flatten images (28x28 -> 784)
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

print("="*50)
print("Training Shape :", x_train.shape)
print("Testing Shape  :", x_test.shape)
print("="*50)



latent_dim = 64

encoder_input = Input(shape=(784,), name="Encoder_Input")

x = Dense(256, activation="relu")(encoder_input)
latent = Dense(latent_dim, activation="relu", name="Latent_Vector")(x)

encoder = Model(
    inputs=encoder_input,
    outputs=latent,
    name="Encoder"
)



decoder_input = Input(shape=(latent_dim,), name="Decoder_Input")

x = Dense(256, activation="relu")(decoder_input)
decoder_output = Dense(784, activation="sigmoid")(x)

decoder = Model(
    inputs=decoder_input,
    outputs=decoder_output,
    name="Decoder"
)

# ============================================================
# Build Autoencoder
# ============================================================

autoencoder_input = encoder_input

encoded = encoder(autoencoder_input)
decoded = decoder(encoded)

autoencoder = Model(
    inputs=autoencoder_input,
    outputs=decoded,
    name="Autoencoder"
)

print("\nAutoencoder Summary\n")
autoencoder.summary()

# ============================================================
# Compile
# ============================================================

autoencoder.compile(
    optimizer="adam",
    loss="binary_crossentropy"
)

# ============================================================
# Print loss after each epoch
# ============================================================

class PrintLoss(tf.keras.callbacks.Callback):

    def on_epoch_end(self, epoch, logs=None):

        print(f"\nEpoch {epoch+1}")
        print(f"Training Loss   : {logs['loss']:.6f}")
        print(f"Validation Loss : {logs['val_loss']:.6f}")
        print("-"*40)

# ============================================================
# Train
# ============================================================

history = autoencoder.fit(
    x_train,
    x_train,
    epochs=10,
    batch_size=256,
    shuffle=True,
    validation_data=(x_test, x_test),
    callbacks=[PrintLoss()],
    verbose=0
)

# ============================================================
# Test on unseen images
# ============================================================

num_images = 5

test_images = x_test[:num_images]

# Before Encoder
print("\nOriginal Image Shape:")
print(test_images.shape)

# Encode
latent_vectors = encoder.predict(test_images)

print("\nLatent Vector Shape:")
print(latent_vectors.shape)

# Decode
reconstructed_images = decoder.predict(latent_vectors)

print("\nReconstructed Shape:")
print(reconstructed_images.shape)

# ============================================================
# Print latent vectors
# ============================================================

for i in range(num_images):

    print("\n"+"="*60)
    print(f"Image {i+1}")
    print("="*60)

    print("Latent Vector:")

    print(np.round(latent_vectors[i],3))

# ============================================================
# Display Images
# ============================================================

plt.figure(figsize=(15,6))

for i in range(num_images):

    # Original Image
    ax = plt.subplot(2, num_images, i+1)
    plt.imshow(test_images[i].reshape(28,28), cmap="gray")
    plt.title("Before Encoder")
    plt.axis("off")

    # Reconstructed Image
    ax = plt.subplot(2, num_images, i+num_images+1)
    plt.imshow(reconstructed_images[i].reshape(28,28), cmap="gray")
    plt.title("After Decoder")
    plt.axis("off")

plt.tight_layout()
plt.show()

# ============================================================
# Show Original, Latent and Reconstruction Together
# ============================================================

for i in range(num_images):

    print("\n"+"#"*70)
    print(f"IMAGE {i+1}")
    print("#"*70)

    fig, ax = plt.subplots(1,2, figsize=(6,3))

    ax[0].imshow(test_images[i].reshape(28,28), cmap="gray")
    ax[0].set_title("Before Encoder")
    ax[0].axis("off")

    ax[1].imshow(reconstructed_images[i].reshape(28,28), cmap="gray")
    ax[1].set_title("After Decoder")
    ax[1].axis("off")

    plt.show()

    print("Latent Vector (64 Dimensions):")
    print(np.round(latent_vectors[i],3))

# ============================================================
# Reconstruction Error
# ============================================================

loss = autoencoder.evaluate(x_test, x_test, verbose=0)

print("\n"+"="*50)
print("Final Reconstruction Loss :", loss)
print("="*50)
```

```python
# ============================================================
# Test on unseen images (0-9)
# ============================================================

(_, _), (x_test_original, y_test) = tf.keras.datasets.mnist.load_data()


x_test = x_test_original.astype("float32") / 255.0
x_test = x_test.reshape(-1, 784)

# اختيار صورة واحدة من كل رقم (0-9)
selected_images = []
selected_labels = []

for digit in range(10):
    idx = np.where(y_test == digit)[0][0]
    selected_images.append(x_test[idx])
    selected_labels.append(digit)

selected_images = np.array(selected_images)

# ============================================================
# Encode & Decode
# ============================================================

latent_vectors = encoder.predict(selected_images, verbose=0)
reconstructed_images = decoder.predict(latent_vectors, verbose=0)

# ============================================================
# Display Results
# ============================================================

plt.figure(figsize=(20,5))

for i in range(10):

    # Original Image
    plt.subplot(2,10,i+1)
    plt.imshow(selected_images[i].reshape(28,28), cmap="gray")
    plt.title(f"Original\n{selected_labels[i]}")
    plt.axis("off")

    # Reconstructed Image
    plt.subplot(2,10,i+11)
    plt.imshow(reconstructed_images[i].reshape(28,28), cmap="gray")
    plt.title(f"Decoded\n{selected_labels[i]}")
    plt.axis("off")

plt.tight_layout()
plt.show()

# ============================================================
# Print Latent Vectors
# ============================================================

for i in range(10):

    print("="*80)
    print(f"Digit: {selected_labels[i]}")
    print("="*80)

    print("\nLatent Vector Shape:", latent_vectors[i].shape)
    print(np.round(latent_vectors[i],3))
    print()
```
