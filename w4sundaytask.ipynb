```python
import numpy as np
```

```python
patients_data = np.array([ 
[45, 80, 200, 120], 
[60, 90, 240, 140], 
[30, 70, 180, 110], 
[50, 85, 220, 130], 
[65, 95, 260, 150] 
])
```

```python

#Q1
print("Shape:", patients_data.shape)
print("Number of patients:", patients_data.shape[0])
print("Number of features:", patients_data.shape[1])
```

```python
# Q2
print("First patient:", patients_data[0])
print("All ages:", patients_data[:, 0])
print("All blood pressure:", patients_data[:, 3])
```

```python
# Q3
print("Average age:", np.mean(patients_data[:, 0]))
print("Max heart rate:", np.max(patients_data[:, 1]))
print("Min cholesterol:", np.min(patients_data[:, 2]))
```

```python
# Part 2: Filtering & Conditions
# Q4
high = patients_data[patients_data[:, 3] > 130]
print("High BP patients:", high)
```

```python
# Q5
highchol = patients_data[patients_data[:, 2] >= 240]
print("High Cholesterol patients:", highchol)
```

```python
# Q6
older = patients_data[patients_data[:, 0] > 50]
print("Patients older than 50:", older)
```

```python
# Part 3: Thinking Like a Data Scientist
# Q7
highrisk = patients_data[
    (patients_data[:, 3] > 140) |
    (patients_data[:, 2] > 240)
]
highrisk
```

```python
# Q8
under50 = patients_data[patients_data[:, 0] < 50]
above50 = patients_data[patients_data[:, 0] >= 50]
print("Avrage HR < 50:", np.mean(under50[:, 1]))
print("Avrage HR >= 50:", np.mean(above50[:, 1]))
```

```python
# Q9
highestbp = np.max(patients_data[:, 3])
lowesthr = np.min(patients_data[:, 1])
print("Highest BP patient:", highestbp)
print("Lowest HR patient:", lowesthr)
```

```python
# Part 4: 
# Q10
ages = patients_data[:, 0]
bp = patients_data[:, 3]
print("ages:", ages)
print("Blood Pressure:", bp)
print("older patient tend to have higher presssure in this data")
```
